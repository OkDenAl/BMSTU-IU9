```julia
using Gaston
t = 0:0.01:1
plot(t, sin.(2π*5*t))
```

```julia
using PyPlot
# use x = linspace(0,2*pi,1000) in Julia 0.6
x = range(0; stop=2*pi, length=1000); y = sin.(3 * x + 4 * cos.(2 * x));
PyPlot.plot(x, y, color="red", linewidth=2.0, linestyle="--")
title("A sinusoidally modulated sinusoid")
```

```julia
using PyPlot

a=2; b=4

x = range(-10, stop=10, length=5000)
y = range(-10, stop=10, length=5000)

z=@. (x^4)/a^4+(y/b)'^4-(x'^2*y^2)/(a*b)
surf(x, y, z,alpha = 0.9, cmap="inferno")
PyPlot.display_figs()
```

```julia
using PyPlot

x = range(-10, stop=10, length=5000)
y = range(-10, stop=10, length=5000)

z=@. x^2+y'^2
surf(x, y, z, alpha = 0.9, cmap="viridis")
PyPlot.display_figs()
```

```julia
using PyPlot

x = range(-10, stop=10, length=5000)
y = range(-10, stop=10, length=5000)

z=@. 1/(1+x'^2) + 1/(1+y^2)
surf(x, y, z, alpha = 0.9, cmap="viridis")
PyPlot.display_figs()
```
