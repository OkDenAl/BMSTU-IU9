### Функции активации

```python
import math

def sigmoid(x): 
    return 1/(1+math.e**(-x))
def sigmoid_der(x): 
    return sigmoid(x)*(1-sigmoid(x))

def tanh(x):
    return (math.e**(2*x) - 1) / (math.e**(2*x) + 1)
def tanh_der(x): 
    return 1 - tanh(x)**2

def relu(x):
    return x if x>0 else 0
def relu_der(x): 
    return 1 if x > 0 else 0

def softmax(arr):
    exp_arr = [(math.e**x) for x in arr]
    sum_exp = sum(exp_arr)
    softmax_arr = [exp / sum_exp for exp in exp_arr]
    return softmax_arr

def softmax_der(arr):
    return [a*b for a,b in zip(softmax(arr),([1 - x for x in softmax(arr)]))]

def linear(x): 
    return x
def linear_der(x): 
    return 1
```

### Функции потерь

```python
def mse_loss(predictions, labels):
    errors = [(label-prediction) ** 2 for label, prediction in zip(labels, predictions)]
    return sum(errors) / len(labels)


def der_mse(predictions, labels):
    res = [(label-prediction) * (-1) for label, prediction in zip(labels, predictions)]
    return res


def cross_entropy(target, pred):
    # Ограничиваем предсказания
    pred = [max(min(p, 1 - 1e-8), 1e-8) for p in pred]
    
    # Рассчитываем кросс-энтропию
    return -sum(t * math.log(p) for t, p in zip(target, pred)) / len(target)

def der_cross_entropy(target, pred):
    # Ограничиваем предсказания
    pred = [max(min(p, 1 - 1e-8), 1e-8) for p in pred]
    
    # Рассчитываем производную кросс-энтропии
    res = [p - t for p, t in zip(pred, target)]
    
    # Нормируем результат
    norm = math.sqrt(sum(r**2 for r in res))
    
    # Возвращаем производную
    return [r / norm for r in res]
```

### Однослойный персептрон

```python
from tqdm import tqdm
from matplotlib import pyplot as plt
import random


class Perceptron:
    def __init__(self, input_size, output_size,loss=mse_loss, activation=linear,activation_der=linear_der,epochs=1000, learning_rate=0.1):
        self.weights = [[random.uniform(-1, 1) for _ in range(input_size)] for _ in range(output_size)]
        self.biases = [random.uniform(-1, 1) for _ in range(output_size)]
        self.learning_rate = learning_rate
        self.activation = activation
        self.activation_der = activation_der
        self.loss=loss
        self.total_loss = []
        self.total_acc = []
        self.epochs = epochs

    def predict(self, inputs):
        outputs = []
        for weights, bias in zip(self.weights, self.biases):
            z = sum(w * i for w, i in zip(weights, inputs)) + bias
            outputs.append(self.activation(z))
        
    
        return outputs

    def train(self, training_inputs, labels):
        average_losses=[]
        accuracy = []
        for _ in tqdm(range(self.epochs)):
            total_loss = 0
            total_accuracy = 0
            for inputs, label in zip(training_inputs, labels):
                predictions = self.predict(inputs)

                loss = self.loss(predictions, label)
                total_loss += loss
                total_accuracy += get_index_of_max_element(predictions) == get_index_of_max_element(label)

                # errors = [(lb-prediction) for lb, prediction in zip(label, predictions)]
                errors = der_mse(predictions, label)
                
                predictions = softmax_der(predictions)
                for i in range(len(self.weights)):
                    for j in range(len(self.weights[i])):
                        gradient = errors[i] * self.activation_der(predictions[i])
                        self.weights[i][j] -= self.learning_rate * gradient * inputs[j]
                    
            accuracy.append(total_accuracy/len(training_inputs))
            average_losses.append(total_loss / len(training_inputs))
            
        self.total_loss.append(average_losses)
        self.total_acc.append(accuracy)
    def show_graphics(self):
        plt.plot(list(range(1, self.epochs+1)), self.total_loss[0],label="Train Loss")
        plt.plot(list(range(1, self.epochs+1)), self.total_loss[1],label="Test Loss")
        plt.grid()
        plt.xlabel("Epochs")
        plt.ylabel("Error")
        plt.legend()
        plt.show()
        
        plt.plot(list(range(1, self.epochs+1)), self.total_acc[0],label="Train Acc")
        plt.plot(list(range(1, self.epochs+1)), self.total_acc[1],label="Test Acc")
        plt.grid()
        plt.legend()
        plt.xlabel("Epochs")
        plt.ylabel("Accuracy")
        plt.show()
```

```python
def get_index_of_max_element(arr):
    max_value = max(arr)  # Находим максимальное значение
    max_index = arr.index(max_value)  # Находим индекс максимального значения
    return max_index
```
