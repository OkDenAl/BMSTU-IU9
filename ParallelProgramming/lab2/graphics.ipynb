```python
import re
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
def parse_file(f):
    strings = f.read().split('\n')
    title, *strings = strings
    data = np.array(list(map(lambda x: [int(x.split()[0]), float(x.split()[1])], filter(bool, strings))))
    data = data.T
    return data,title
```

```python
def draw_file(f):
    data,title = parse_file(f)
    plt.title(title)
    plt.plot(data[0], data[1])

    plt.xlabel('Count of proccess', fontsize=12)
    plt.ylabel('Time', fontsize=12)
    plt.legend(fontsize=10)
    plt.grid()


    plt.show()
```

```python
files = os.listdir()
for file in files:
    if re.fullmatch('.*\.txt', file):
        with open(file, 'r') as f:
            draw_file(f)
```
