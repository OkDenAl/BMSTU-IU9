# 2. Толкования в словарях

1) **Словарь Даля:** Круглый, открытый сверху, защипанный только с краев пирожок или лепешка, обыкновенно с творогом
2) **Словарь Ушакова:** Толстая лепешка с загнутыми краями и с покрытой творогом серединой.
3) **Словарь Ожегова:** Род открытого пирога, пирожка с творожной начинкой.
4) **Словарь Кузнецова:** Открытый пирог или пирожок с загнутыми кверху краями и начинкой обычно из творога (реже — джема, ягод, фруктов).


**Сходства:**
1) Везде ватрушка с творожной начинкой
2) В большинстве определений идёт акцент на том, что края должны быть защипаны

**Различия:**
1) Каждый автор делает разные акценты на форме, краях ватрушки
2) В словаре Кузнецова указано, что ватрушка необязательно из творога
3) Везде кроме Ушакова и Даля, ватрушка это пирожок или пирог, у них это характеризуется как лепешка

Как за нашим за двором висит ватрушка с творогом.

    2.2) В. А. Соллогуб. Сережа (1838) 
В деревне молоко, природа, сметана, чистая любовь у ручейка, за обедом ватрушки — жизнь патриархальная.
3) **В газетном корпусе:**
   3.1) Жириновский предлагает лишать прав за привязанные к автомобилям «ватрушки» // Парламентская газета, 2021.01.11 
Катание на санках, тюбингах или «ватрушках», привязанных к автомобилю, нужно запретить, за такие действия необходимо лишать водительских прав, написал 11 января лидер ЛДПР Владимир Жириновский в своём Telegram-канале.
    3.2) Жительница Нижегородской области насмерть разбилась на тюбинге // РИА Новости, 09.01.2020 
По предварительным данным, 1 января 2020 года компания молодых людей пришла покататься на "ватрушке" в одном из оврагов около Павлово.
4) **Употребление этого слова в Яндекс-новостях:**
    4.1) [Гигантская Ватрушка](https://izvmor.ru/novosti/obshchestvo/v-mordovii-ispekli-vatrushku-vesom-250-kg/?utm_source=yxnews&utm_medium=desktop&utm_referrer=https%3A%2F%2Fdzen.ru%2Fnews%2Fsearch)
    4.2) [Покататься на ватрушке](https://ura.news/news/1052926077?utm_source=yxnews&utm_medium=desktop&utm_referrer=https%3A%2F%2Fdzen.ru%2Fnews%2Fsearch)


**Сходства:**
1) Во всех случаях слово "ватрушка" остаётся существительным женского рода.
2) В любом значении сохраняется ассоциация с круглой формой (пирожок → круглый, тюбинг → круглый).
3) В разных корпусах слово используется в бытовом, разговорном контексте (еда, развлечения, простые житейские ситуации).

**Различия:**
1) Меняется значение слова "ватрушка", оно всё чаще начинает использоваться в контексте тюбинга (особенно в газетном корпусе)

# 5. Выводы

В ходе исследования слова было выявелно **следующее:**
1) Слово «ватрушка» расширило значение — от традиционной выпечки к обозначению предмета досуга (тюбинг).
2) Причина появления нового значения - перенос по признаку формы (круглая ватрушка с краями и круглый надувной тюбинг с краями).
3) Слово ватрушка всё ещё употребимо и в обычном контексте (блюдо)
