```python
import re

text = open('text.txt').read()

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^а-яё\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

clean_text = preprocess_text(text)
```

```python
import pymorphy3
morph_analyzer = pymorphy3.MorphAnalyzer()

def get_lemmas_pymorphy(text):
    words = text.split()
    lemmas = []
    for word in words:
        parsed = morph_analyzer.parse(word)
        if parsed:
            lemma = parsed[0].normal_form
            lemmas.append(lemma)
    return lemmas

lemmas_pymorphy = get_lemmas_pymorphy(clean_text)
print("Леммы PyMorphy (первые 50):")
print(lemmas_pymorphy[:50])
```

```python
import stanza
stanza.download('ru')
nlp = stanza.Pipeline('ru', processors='tokenize,lemma,pos')

def get_lemmas_stanza(text):
    doc = nlp(text)
    lemmas = []
    for sentence in doc.sentences:
        for word in sentence.words:
            lemmas.append(word.lemma)
    return lemmas

lemmas_stanza = get_lemmas_stanza(clean_text)
print("\nЛеммы Stanza (первые 50):")
print(lemmas_stanza[:50])
```

```python
from collections import Counter


def build_lemma_frequency_corpus(text, lemma_func):
    lemma_counter = Counter()
    
    clean_text = preprocess_text(text)
    lemmas = lemma_func(clean_text)
    lemma_counter.update(lemmas)
    
    return lemma_counter
```
